```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# plot the csv data from example.csv as x,y scatter plot
perlinData = pd.read_csv('perlin1D.csv')


plt.plot(perlinData['x'], perlinData['y'], color='red', label='perlin data')

plt.legend()


plt.xlabel('x')
plt.ylabel('y')
plt.title('x,y scatter plot')
plt.show()
```

```python
# create a 3d plot of the data from perlin.csv as a surface plot
perlinData = pd.read_csv('perlin.csv')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = perlinData['x']
y = perlinData['y']
z = perlinData['z']

ax.plot_trisurf(x, y, z, cmap='viridis')
plt.show()
```

```python
# plot 3d data as 2d plot with color as the 3rd dimension
perlinData = pd.read_csv('perlin.csv')

x = perlinData['x']
y = perlinData['y']
z = perlinData['z']



plt.scatter(perlinData['x'], perlinData['y'], c=perlinData['z'], cmap='viridis')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('x,y scatter plot')
plt.show()
```
